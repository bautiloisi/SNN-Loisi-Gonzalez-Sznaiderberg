```python
data = pd.read_csv("/content/autos.csv")

# Codificar las columnas categóricas utilizando Label Encoding
label_encoder = LabelEncoder()

# Iterar sobre las columnas y aplicar Label Encoding a las categóricas
for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = label_encoder.fit_transform(data[column])
data
```

```python
def sigmoid(x):
    return 1 / ( 1 + np.exp(-x) )


def sigmoid_derivative(x):
    return x * (1 - x)
```

Se define la función de activación sigmoidal en las funciones sigmoid(x) y sigmoid_derivative(x). La función sigmoid(x) se usa para calcular las salidas de la red neuronal y sigmoid_derivative(x) se usa en el cálculo de ajustes de pesos durante el entrenamiento.

```python
training_inputs = data.drop(columns=['Sportscar','Luxurious','Family','SUV'])

training_outputs = data[['Sportscar','Luxurious','Family','SUV']]

display(training_inputs)
display(training_outputs)
```

```python
n = training_inputs.shape[1]
k = training_outputs.shape[1]
print(n, k)
synaptic_weights = np.random.random((n, k))

synaptic_weights
```

```python
eval_NN = lambda x, w: sigmoid(np.dot(x, w))
```

```python
# 50000 epochs
for iteration in range(5000):

    # Arranca con los valores de training
    input_layer = training_inputs

    # Genera un output haciendo la suma de los valores de input multiplicados por su peso sinaptico, y aplicandole la función sigmoide para normalizar a -1, 1
    outputs = eval_NN(input_layer, synaptic_weights)

    # Calculo el error
    error = training_outputs - outputs
    #display(error)
    # Ajusto los pesos en funcion al error y al resultado obtenido
    adjustments = error * sigmoid_derivative(outputs)
    synaptic_weights += np.dot(input_layer.T, adjustments)
```

El proceso de forward propagation se realiza en el bucle for iteration in range(5000) donde se calculan las salidas de la red neuronal (outputs) mediante la función eval_NN(input_layer, synaptic_weights). Aquí, se multiplican las entradas por los pesos sinápticos y se aplica la función sigmoidal para normalizar los resultados.

El proceso de back propagation se realiza en las siguientes líneas:
Se calcula el error entre las salidas deseadas (training_outputs) y las salidas de la red (outputs) en error = training_outputs - outputs.
Luego, se calculan los ajustes de pesos (adjustments) utilizando el error y la derivada de la función sigmoidal en adjustments = error * sigmoid_derivative(outputs).
Finalmente, se actualizan los pesos sinápticos en synaptic_weights += np.dot(input_layer.T, adjustments).
El gradiente descendente se aplica indirectamente durante el proceso de back propagation. Los pesos sinápticos se ajustan, lo que es un paso esencial en el entrenamiento de la red neuronal para minimizar la función de pérdida.
